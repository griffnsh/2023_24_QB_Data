```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'http://pfref.com/pi/share/WNCsx'

page = requests.get(url)

data = BeautifulSoup(page.text, 'html')
```

```python
# data.find('table')
```

```python
table = data.find('table')
```

```python
stat_categories = data.find_all('th')
# print(stat_categories)
```

```python
stat_categories_titles = [title.text.strip() for title in stat_categories]
print(stat_categories_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = stat_categories_titles)

df
```

```python
column_data = table.find_all('tr')[1:]
```

```python
for row in column_data:
    row_data = row.find_all('td')
    player_row_data = [data.text.strip() for data in row_data]
    print(player_row_data)
    
    length = len(df)
    df.loc[length] = player_row_data
    
```
