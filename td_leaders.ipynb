```python
from bs4 import BeautifulSoup
import requests
```

```python
# url = 'https://www.pro-football-reference.com/years/2023/passing.htm#passing::pass_yds'

url = 'http://pfref.com/pi/share/MY6d8'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
# soup.find('table')
```

```python
table = soup.find('table')
print(table)
```

```python
qb_categories = soup.find_all('th')
```

```python
# qb_categories
```

```python
qb_categories_titles = [title.text.strip() for title in qb_categories]

print(qb_categories_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = qb_categories_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    qb_row_data = [data.text.strip() for data in row_data]
    print(qb_row_data)

    length = len(df)
    df.loc[length] = qb_row_data
```
